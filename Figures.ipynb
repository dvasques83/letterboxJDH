```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Holdings

```python
df = pd.read_csv('holdings01.csv', delimiter=',')
```

```python
years = list(df['Year'])
frequency= list(df['Count'])
```

```python
# figure environment
fig, ax = plt.subplots(figsize=(4,3), dpi=500)

ax.plot(years, frequency, c='#9ecae1', markersize=2, marker='o', linestyle='-')

# axis set-up
ax.set_xlabel('Year', fontsize=7)
ax.set_ylabel('Number of holdings', fontsize=7)
ax.tick_params(axis='both', which='major', labelsize=7) 

ax.set_ylim(-0.00, 1275)
ax.set_xlim(1959, 1991)

#Set legend
#ax.legend(loc='lower right', frameon=False, numpoints=1, fontsize=10)

# Layout
plt.tight_layout()
#plt.legend(loc='lower right', fontsize=16)
#ax.tick_params(axis='x', pad=10)

#plt.title('')
plt.savefig(f'media/holdings-evolution.png', format='png', dpi=500)
plt.show()
```

### Panama

```python
df = pd.read_csv('panama.csv', delimiter=',', encoding='latin-1')
```

```python
years = list(df['année'])
frequency= list(df['number of companies'])
percentage = list(df['percentage'])
```

```python
# figure environment
fig, ax1 = plt.subplots(figsize=(4,3), dpi=500)

ax1.plot(years, frequency, c='#9ecae1', markersize=2, marker='o', linestyle='-', label='Number of companies')
ax2 = ax1.twinx()
ax2.plot(years, percentage, c='#f4a582', markersize=2, marker='o', linestyle='-', label= 'Percentage')

# axis set-up
ax1.set_xlabel('Year', fontsize=7)
ax1.set_ylabel('Number of companies', fontsize=7)
ax2.set_ylabel('Percentage', fontsize=7) 

ax1.tick_params(axis='both', which='major', labelsize=7) 
ax2.tick_params(axis='y', which='major', labelsize=7)

ax1.set_ylim(-50, 820)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.2%}'.format(y))) 

#Set legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right', frameon=False, fontsize=7)

# Layout
plt.tight_layout()
#plt.legend(loc='lower right', fontsize=16)
#ax.tick_params(axis='x', pad=10)

#plt.title('')
#plt.savefig(f'media/panama-evolution.png', format='png', dpi=500)
plt.show()
```
